```python
tmp=list()
for specie in df['species']:
  if specie =='Iris-setosa':
    tmp.append(0)
  elif specie=='Iris-versicolor':
    tmp.append(1)
  else:
    tmp.append(2)

df['target']=tmp

print(df['target'].unique())
```

```python
df.head(5)
```

```python
x=df.drop(['species','target'],axis=1)
x.head(5)
```

```python
y=df['target']
y.head(3)
```

```python
print(type(x),type(y))
```

```python
import numpy as np
x=x.to_numpy()[:,(2,3)]
type(y)
```

```python
#SPLIT THE DATA  INTO TRAIN AND TEST DATASETS
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.8,random_state=45)

#CREATE THE MODEL CLASSIFICATION
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
#NOW PREDICT THE RESULT USING PREDICT METHOD
yPredictions=model.predict(xtest)
yPredictions
```

```python
labels=df['species'].unique()
print(labels)
print(labels.shape)
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
label_encoder.fit(labels)

for i, item in enumerate(label_encoder.classes_):
  print(item,'---->',i)
```

```python
print(yPredictions[:10])
encodedLabels=label_encoder.inverse_transform(yPredictions)
print(encodedLabels[:10])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(ytest,yPredictions))
```

```python
print(accuracy_score(ytest,yPredictions))
#We can see that accuracy of the model is 97.37 percent which is very accurate.
```
